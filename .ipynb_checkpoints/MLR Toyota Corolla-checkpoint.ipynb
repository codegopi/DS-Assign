```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data = pd.read_csv("ToyotaCorolla.csv", encoding="latin1")
```

```python
data.head()
```

```python
data.info()
```

```python
# selecting only the relevant columns

data1 = data[["Price","Age_08_04","KM","HP", "cc","Doors","Gears","Quarterly_Tax","Weight"]]
data1.head()
```

```python
data2 = data1.rename({"Age_08_04": "Age","Quarterly_Tax": "Tax"}, axis=1)
data2.head()
```

```python
data2[data2.duplicated()]
```

```python
data3 = data2.drop_duplicates().reset_index(drop=True)
data3.describe()
```

```python
data3.corr()
```

```python
sns.set_style(style="darkgrid")
sns.pairplot(data3)
```

```python
model = smf.ols("Price~Age+KM+HP+cc+Doors+Gears+Weight", data=data3).fit()
```

```python
model.params
```

```python
model.tvalues, np.round(model.pvalues, 4)
```

```python
#model accuracy
model.rsquared, model.rsquared_adj
```

```python
slr_c = smf.ols("Price~cc", data=data3).fit()
slr_c.tvalues, np.round(slr_c.pvalues,4)
```

```python
slr_d = smf.ols("Price~Doors", data=data3).fit()
slr_d.tvalues, np.round(slr_d.pvalues)
```

```python
mlr_cd = smf.ols("Price~cc+Doors", data=data3).fit()
mlr_cd.tvalues, np.round(mlr_cd.pvalues)
```

```python
# checking for vif value of more than 20

rsq_age = smf.ols("Age~KM+cc+Doors+Gears+Tax+Weight", data=data3).fit().rsquared
vif_age=1/(1-rsq_age)

rsq_km = smf.ols("KM~Age+cc+Doors+Gears+Tax+Weight", data=data3).fit().rsquared
vif_km=1/(1-rsq_km)

rsq_cc = smf.ols("cc~KM+Age+Doors+Gears+Tax+Weight", data=data3).fit().rsquared
vif_cc=1/(1-rsq_cc)

rsq_door = smf.ols("Doors~KM+cc+Age+Gears+Tax+Weight", data=data3).fit().rsquared
vif_door=1/(1-rsq_door)

rsq_gear = smf.ols("Gears~KM+cc+Doors+Age+Tax+Weight", data=data3).fit().rsquared
vif_gear=1/(1-rsq_gear)

rsq_tax = smf.ols("Tax~KM+cc+Doors+Age+Gears+Weight", data=data3).fit().rsquared
vif_tax=1/(1-rsq_tax)

rsq_weight = smf.ols("Weight~KM+cc+Doors+Age+Tax+Gears", data=data3).fit().rsquared
vif_weight=1/(1-rsq_weight)

d1 = {"Features": ["Age", "KM", "CC", "Door", "Gear", "Tax", "Weight"],
      "ViF": [vif_age, vif_km, vif_cc, vif_door, vif_gear, vif_tax, vif_weight]
     }
vif_df = pd.DataFrame(d1)
vif_df
```

```python
sm.qqplot(model.resid, line="q")
plt.title("Normal Q-Q Plot of residuals")
plt.show()
```

```python
def std_values(vals):
    return (vals - vals.mean())/vals.std()
```

```python
plt.scatter(std_values(model.fittedvalues), std_values(model.resid))
plt.title("Residual Plot")
plt.xlabel("Standardized Fitted Values")
plt.ylabel("Standardized Residual Values")
plt.show()
```

```python
#cooks distance
(c, _) = model.get_influence().cooks_distance
c
```

```python
data3[data3.index.isin([80])]
```

```python
#improving the model
data_new = data3.copy()
data_new.head()
```

```python
data4 = data_new.drop(data_new.index[[80]], axis=0).reset_index(drop=True)
data4
```

```python
while model.rsquared < 0.90:
    for c in [np.max(c) > 0.5]:
        model=smf.ols("Price~Age+KM+cc+HP+Doors+Gears+Weight", data=data4).fit()
        (c,_) = model.get_influence().cooks_distance
        c
        np.argmax(c), np.max(c)
        data4=data4.drop(data4.index[[np.argmax(c)]], axis=0).reset_index(drop=True)
        data4
    else:
        final_model=smf.ols("Price~Age+KM+cc+HP+Doors+Gears+Weight", data=data4).fit()
        final_model.rsquared, final_model.aic
        print("The model accuracy is improved to ", final_model.rsquared )
```
