```python
data1 = pd.get_dummies(data, columns=["job","marital","education","contact","poutcome"])
pd.set_option("display.max.columns", None)
data1
```

```python
data1["default"] = np.where(data1["default"].str.contains("yes"),1,0)
data1["housing"] = np.where(data1["housing"].str.contains("yes"),1,0)
data1["loan"] = np.where(data1["loan"].str.contains("yes"),1,0)
data1["y"] = np.where(data1["y"].str.contains("yes"),1,0)
data1
```

```python
data1["month"].value_counts()
```

```python
month = {"month" : {"jan":1,"feb":2,"mar":3,"apr":4,"may":5,"jun":6,"jul":7,"aug":8,"sep":9,"oct":10,"nov":11,"dec":12}}
data1 = data1.replace(month)
```

# Model Building

```python
x = pd.concat([data1.iloc[:,0:11],data1.iloc[:,12:]], axis=1)
y = data1.iloc[:,11]
```

```python
classifier = LogisticRegression()
classifier.fit(x,y)
```

# Model Prediction

```python
y_pred = classifier.predict(x)
y_pred
```

```python
y_pred_df = pd.DataFrame({"Actual_y":y, "Predicted_y": y})
y_pred_df
```
