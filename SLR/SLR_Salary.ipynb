```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
regressor = pd.read_csv("Salary_Data.csv")
regressor.head()
```

```python
X = regressor["YearsExperience"].values.reshape(-1,1)
y = regressor["Salary"].values.reshape(-1,1)
```

```python
lm = LinearRegression()
lm.fit(X, y)
lm.predict(X)
```

```python
plt.scatter(X,y, color="Red", label="Actual")
plt.plot(X, lm.predict(X), color="Blue", label="Predicted")
plt.show()
```

```python
first_model = smf.ols("Salary~YearsExperience", data=regressor).fit()
first_model.summary()
```

```python
second_model = smf.ols("Salary~np.log(YearsExperience)", data=regressor).fit()
second_model.summary()
```

```python
third_model = smf.ols("np.log(Salary)~YearsExperience", data=regressor).fit()
third_model.summary()
```

```python
regressor["Exp_Sq"] = regressor["YearsExperience"]**2
forth_model = smf.ols("np.log(Salary)~YearsExperience+Exp_Sq", data=regressor).fit()
forth_model.summary()
```
